```python
import seaborn as sns
import matplotlib.pyplot as plt # Seaborn usa Matplotlib por baixo dos panos
import pandas as pd
import numpy as np
```

Melhorando a Estética com Seaborn

```python
# Carregando um dataset de exemplo do Seaborn
# 'tips' é um dataset sobre gorjetas em restaurantes
df_tips = sns.load_dataset('tips')
print("DataFrame 'tips' (primeiras 5 linhas):\n", df_tips.head())

# Plotando um histograma com Seaborn (já vem com estilo melhor)
sns.histplot(data=df_tips, x='total_bill', kde=True) # kde=True para adicionar a linha de densidade
plt.title("Distribuição da Conta Total")
plt.xlabel("Conta Total ($)")
plt.ylabel("Frequência")
plt.show()
```

Gráfico de Dispersão Avançado (sns.scatterplot() ou sns.relplot()):

```python
# Relação entre conta total e gorjeta, diferenciando por gênero
sns.scatterplot(data=df_tips, x='total_bill', y='tip', hue='sex', size='size', sizes=(20, 400), alpha=0.7)
plt.title("Conta Total vs. Gorjeta por Gênero")
plt.xlabel("Conta Total ($)")
plt.ylabel("Gorjeta ($)")
plt.show()
```

Gráfico de Barras com Agregação (sns.barplot()):

```python
# Média da conta total por dia da semana
sns.barplot(data=df_tips, x='day', y='total_bill', estimator=np.mean, errorbar='sd') # errorbar='sd' mostra desvio padrão
plt.title("Média da Conta Total por Dia da Semana")
plt.xlabel("Dia da Semana")
plt.ylabel("Média da Conta Total ($)")
plt.show()
```

Gráfico de Contagem (sns.countplot()):

```python
sns.countplot(data=df_tips, x='day', hue='time') # Contagem de refeições por dia e período
plt.title("Contagem de Refeições por Dia e Período")
plt.xlabel("Dia da Semana")
plt.ylabel("Número de Refeições")
plt.show()
```

Box Plot (sns.boxplot()):

```python
sns.boxplot(data=df_tips, x='day', y='tip', hue='smoker')
plt.title("Distribuição de Gorjetas por Dia e Fumante")
plt.xlabel("Dia da Semana")
plt.ylabel("Gorjeta ($)")
plt.show()
```

Heatmap (sns.heatmap()):

```python
# Calculando a matriz de correlação (apenas para colunas numéricas)
correlation_matrix = df_tips[['total_bill', 'tip', 'size']].corr()

plt.figure(figsize=(6, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Matriz de Correlação das Gorjetas")
plt.show()
```

Personalização e Combinação de Gráficos

```python
# Combinando tipos de gráficos e personalização
plt.figure(figsize=(12, 6))

# Gráfico de dispersão
sns.scatterplot(data=df_tips, x='total_bill', y='tip', hue='day', size='size', sizes=(50, 500), alpha=0.7)

# Adicionando uma linha de regressão (sem intervalo de confiança para clareza)
sns.regplot(data=df_tips, x='total_bill', y='tip', scatter=False, color='red', ci=None)

plt.title("Relação entre Conta Total e Gorjeta com Linha de Tendência", fontsize=16)
plt.xlabel("Conta Total ($)", fontsize=12)
plt.ylabel("Gorjeta ($)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Dia da Semana', bbox_to_anchor=(1.05, 1), loc='upper left') # Posiciona a legenda fora
plt.tight_layout() # Ajusta para evitar sobreposição de elementos
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Para mostrar os gráficos dentro do notebook
%matplotlib inline
```

```python
# Criando o DataFrame clientes

df_clientes = pd.DataFrame({
    "ID": [1, 2, 3, 4, 5],
    "Nome": ["Ana", "Bruno", "Carla", "Diego", "Elaine"],
    "Idade": [28, 35, 22, 40, 30],
    "Cidade": ["São Paulo", "Rio de Janeiro", "Belo Horizonte", "Curitiba", "Fortaleza"],
    "Renda": [1500.50, 2300.00, 1800.75, 1200.00, 3000.30]
})
```

```python
# Exercício 1:

# Criando o DataFrame produtos
produtos = pd.DataFrame({
    'Item': ['Laptop', 'Mouse', 'Teclado', 'Monitor'],
    'Quantidade': [25, 60, 40, 30]
})

# Gráfico de barras
plt.figure(figsize=(8, 5))
plt.bar(produtos['Item'], produtos['Quantidade'], color='green')
plt.title('Quantidade de Produtos em Estoque')
plt.xlabel('Produto')
plt.ylabel('Quantidade')
plt.show()
```

```python
# Exercício 2:

# Criando os arrays
dias = np.arange(1, 8)
temperaturas = np.array([25, 27, 26, 28, 29, 27, 26])

# Gráfico de linha
plt.figure(figsize=(8, 5))
plt.plot(dias, temperaturas, linestyle=':', marker='o', color='blue')
plt.title('Variação de Temperatura Semanal')
plt.xlabel('Dia')
plt.ylabel('Temperatura (°C)')
plt.grid(True)
plt.show()
```

```python
# Exercício 3:

# Histograma da idade
plt.figure(figsize=(8, 5))
sns.histplot(df_clientes['Idade'], bins=5, kde=False, color='skyblue')
plt.title('Distribuição de Idades dos Clientes')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

```python
# Exercício 4:

df_clientes['genero'] = ['Feminino', 'Masculino', 'Feminino', 'Masculino', 'Feminino']

# Gráfico de dispersão
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_clientes, x='Idade', y='Renda', hue='genero', palette='Set2')
plt.title('Renda vs Idade dos Clientes')
plt.xlabel('Idade')
plt.ylabel('Renda (R$)')
plt.legend(title='Gênero')
plt.show()
```

```python
# Exercício 5:

# Boxplot da renda por gênero
plt.figure(figsize=(8, 5))
sns.boxplot(data=df_clientes, x='genero', y='Renda', palette='pastel')
plt.title('Distribuição de Renda por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Renda (R$)')
plt.show()
```
