Introdução ao Matplotlib e pyplot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd # Pode ser útil para criar dados para visualização
```

Configurações Básicas de um Plot

```python
# Criando dados simples
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 6])

# Criando um plot básico
plt.plot(x, y) # Plota y em função de x

# Adicionando título e rótulos aos eixos
plt.title("Meu Primeiro Gráfico de Linha")
plt.xlabel("Eixo X (Variável Independente)")
plt.ylabel("Eixo Y (Variável Dependente)")

# Exibindo o gráfico
plt.show()
```

Gráfico de Linha (plt.plot()):

```python
# Dados de vendas ao longo dos meses
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun']
vendas = [100, 120, 150, 130, 180, 200]

plt.plot(meses, vendas, marker='o', linestyle='--', color='blue', label='Vendas Mensais')
plt.title("Vendas ao Longo dos Meses")
plt.xlabel("Mês")
plt.ylabel("Quantidade de Vendas")
plt.grid(True) # Adiciona uma grade ao gráfico
plt.legend() # Mostra a legenda (se houver rótulo)
plt.show()
```

Gráfico de Barras (plt.bar()):

```python
categorias = ['A', 'B', 'C', 'D']
valores = [25, 40, 30, 55]

plt.bar(categorias, valores, color=['red', 'green', 'blue', 'orange'])
plt.title("Comparação de Valores por Categoria")
plt.xlabel("Categoria")
plt.ylabel("Valor")
plt.show()
```

Histograma (plt.hist()):

```python
# Gerando dados de exemplo com distribuição normal
dados_dist = np.random.normal(loc=50, scale=10, size=1000) # Média 50, Desvio Padrão 10

plt.hist(dados_dist, bins=30, edgecolor='black', alpha=0.7)
plt.title("Distribuição de Dados")
plt.xlabel("Valores")
plt.ylabel("Frequência")
plt.show()
```

Gráfico de Dispersão (plt.scatter()):

```python
# Dados de exemplo: horas de estudo vs. nota
horas_estudo = np.array([2, 3, 5, 4, 7, 6, 8, 3, 5])
notas = np.array([50, 60, 75, 70, 90, 85, 95, 65, 80])

plt.scatter(horas_estudo, notas, color='purple', alpha=0.8, s=100) # s é o tamanho dos pontos
plt.title("Horas de Estudo vs. Notas")
plt.xlabel("Horas de Estudo")
plt.ylabel("Notas")
plt.show()
```

Gráfico de Pizza (plt.pie()):

```python
# Dados de percentuais de mercado
labels = ['Apple', 'Samsung', 'Xiaomi', 'Outros']
sizes = [40, 30, 15, 15]
explode = (0, 0.1, 0, 0) # Destaca a fatia da Samsung

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, explode=explode, shadow=True)
plt.title("Fatias de Mercado de SmartPhones")
plt.axis('equal') # Garante que o círculo seja desenhado como um círculo
plt.show()
```

Personalização Avançada

```python
# Exemplo de Subplots
x = np.linspace(0, 2 * np.pi, 100) # 100 pontos entre 0 e 2*pi
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(10, 5)) # Define o tamanho da figura

plt.subplot(1, 2, 1) # 1 linha, 2 colunas, primeiro plot
plt.plot(x, y_sin, color='red')
plt.title("Função Seno")
plt.xlabel("Ângulo (radianos)")
plt.ylabel("Seno")
plt.grid(True)

plt.subplot(1, 2, 2) # 1 linha, 2 colunas, segundo plot
plt.plot(x, y_cos, color='green')
plt.title("Função Cosseno")
plt.xlabel("Ângulo (radianos)")
plt.ylabel("Cosseno")
plt.grid(True)

plt.tight_layout() # Ajusta automaticamente os parâmetros de subplot para que caibam na área da figura
plt.show()
```
