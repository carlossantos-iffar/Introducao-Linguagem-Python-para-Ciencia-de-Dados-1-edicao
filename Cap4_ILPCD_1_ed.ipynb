Listas (list)

```python
# Criando listas
lista_vazia = []
numeros = [1, 2, 3, 4, 5]
frutas = ["maçã", "banana", "cereja", "banana"]
mista = ["Maria", 25, 1.65, True]

print(f"Lista de números: {numeros}")
print(f"Lista de frutas: {frutas}")
print(f"Lista mista: {mista}")

print(f"Tipo da variável 'numeros': {type(numeros)}")

# Acessando elementos (indexação)
# A contagem de índices em Python começa em 0!
print(f"\nPrimeiro número: {numeros[0]}")     # Saída: Primeiro número: 1
print(f"Terceira fruta: {frutas[2]}")         # Saída: Terceira fruta: cereja
print(f"Último elemento da lista mista: {mista[3]}") # Saída: Último elemento da lista mista: True

# Acessando do final (índices negativos)
print(f"Última fruta: {frutas[-1]}") # Saída: Última fruta: banana
print(f"Penúltimo número: {numeros[-2]}") # Saída: Penúltimo número: 4
```

Fatiamento (Slicing):

```python
numeros = [10, 20, 30, 40, 50, 60, 70, 80, 90]

print(f"\nTodos os números: {numeros[:]}")       # Saída: Todos os números: [10, 20, 30, 40, 50, 60, 70, 80, 90]
print(f"Do índice 2 ao 5 (não incluindo 5): {numeros[2:5]}") # Saída: Do índice 2 ao 5 (não incluindo 5): [30, 40, 50]
print(f"Do início ao índice 4: {numeros[:5]}")     # Saída: Do início ao índice 4: [10, 20, 30, 40, 50]

print(f"Do índice 6 ao final: {numeros[6:]}")     # Saída: Do índice 6 ao final: [70, 80, 90]
print(f"Pular de 2 em 2: {numeros[::2]}")         # Saída: Pular de 2 em 2: [10, 30, 50, 70, 90]
print(f"Lista invertida: {numeros[::-1]}")        # Saída: Lista invertida: [90, 80, 70, 60, 50, 40, 30, 20, 10]
```

Modificando Listas (Mutabilidade):

```python
frutas = ["maçã", "banana", "cereja"]
print(f"\nLista original de frutas: {frutas}") # Saída: Lista original de frutas: ['maçã', 'banana', 'cereja']

# Modificar um elemento
frutas[1] = "laranja"
print(f"Lista após modificar 'banana' para 'laranja': {frutas}") # Saída: Lista após modificar 'banana' para 'laranja': ['maçã', 'laranja', 'cereja']

# Adicionar elementos
frutas.append("uva") # Adiciona no final
print(f"Lista após adicionar 'uva': {frutas}") # Saída: Lista após adicionar 'uva': ['maçã', 'laranja', 'cereja', 'uva']

frutas.insert(1, "kiwi") # Adiciona na posição específica (índice 1)
print(f"Lista após inserir 'kiwi' no índice 1: {frutas}") # Saída: Lista após inserir 'kiwi' no índice 1: ['maçã', 'kiwi', 'laranja', 'cereja', 'uva']

# Remover elementos
frutas.remove("laranja") # Remove a primeira ocorrência do valor
print(f"Lista após remover 'laranja': {frutas}") # Saída: Lista após remover 'laranja': ['maçã', 'kiwi', 'cereja', 'uva']

item_removido = frutas.pop(2) # Remove pelo índice e retorna o item removido
print(f"Lista após remover item no índice 2: {frutas}") # Saída: Lista após remover item no índice 2: ['maçã', 'kiwi', 'uva']
print(f"Item removido com pop(): {item_removido}") # Saída: Item removido com pop(): cereja

del frutas[0] # Remove pelo índice (não retorna nada)
print(f"Lista após remover item no índice 0 com del: {frutas}") # Saída: Lista após remover item no índice 0 com del: ['kiwi', 'uva']

frutas.clear() # Remove todos os elementos da lista
print(f"Lista após limpar todos os elementos: {frutas}") # Saída: Lista após limpar todos os elementos: []
```

```python
numeros = [1, 5, 2, 8, 3, 5]

print(f"\nComprimento da lista (número de itens): {len(numeros)}") # Saída: Comprimento da lista (número de itens): 6
print(f"Contar ocorrências do 5: {numeros.count(5)}") # Saída: Contar ocorrências do 5: 2
print(f"Índice da primeira ocorrência do 8: {numeros.index(8)}") # Saída: Índice da primeira ocorrência do 8: 3

numeros.sort() # Ordena a lista em ordem crescente (modifica a própria lista)
print(f"Lista ordenada: {numeros}") # Saída: Lista ordenada: [1, 2, 3, 5, 5, 8]

numeros.reverse() # Inverte a ordem da lista (modifica a própria lista)
print(f"Lista invertida: {numeros}") # Saída: Lista invertida: [8, 5, 5, 3, 2, 1]
```

Tuplas (tuple)

Dicionários (dict)

```python
# Criando dicionários
dicionario_vazio = {}
aluno = {
    "nome": "Carlos",
    "idade": 28,
    "cidade": "São Paulo",
    "cursos": ["Python", "SQL"]
}
estoque = {"banana": 100, "maçã": 150, "laranja": 75}

print(f"\nDicionário do aluno: {aluno}")
print(f"Dicionário de estoque: {estoque}")
print(f"Tipo da variável 'aluno': {type(aluno)}")

# Acessando valores por chave
print(f"Nome do aluno: {aluno['nome']}") # Saída: Nome do aluno: Carlos
print(f"Quantidade de maçãs no estoque: {estoque['maçã']}") # Saída: Quantidade de maçãs no estoque: 150

# Método get() para evitar erros se a chave não existir
print(f"Cidade do aluno (get): {aluno.get('cidade')}") # Saída: Cidade do aluno (get): São Paulo
print(f"Curso inexistente (get): {aluno.get('curso_inexistente', 'Não informado')}") # Saída: Curso inexistente (get): Não informado
```

Modificando Dicionários:

```python
aluno = {
    "nome": "Carlos",
    "idade": 28,
    "cidade": "São Paulo"
}
print(f"\nDicionário original do aluno: {aluno}")

# Adicionar um novo par chave:valor
aluno["email"] = "carlos@email.com"
print(f"Dicionário após adicionar email: {aluno}") # Saída: Dicionário após adicionar email: {'nome': 'Carlos', 'idade': 28, 'cidade': 'São Paulo', 'email': 'carlos@email.com'}

# Modificar um valor existente
aluno["idade"] = 29
print(f"Dicionário após atualizar idade: {aluno}") # Saída: Dicionário após atualizar idade: {'nome': 'Carlos', 'idade': 29, 'cidade': 'São Paulo', 'email': 'carlos@email.com'}

# Remover um par chave:valor
del aluno["cidade"]
print(f"Dicionário após remover cidade: {aluno}") # Saída: Dicionário após remover cidade: {'nome': 'Carlos', 'idade': 29, 'email': 'carlos@email.com'}

# Outro método para remover e obter o valor
email_removido = aluno.pop("email")
print(f"Dicionário após remover email com pop(): {aluno}") # Saída: Dicionário após remover email com pop(): {'nome': 'Carlos', 'idade': 29}
print(f"Email removido: {email_removido}") # Saída: Email removido: carlos@email.com
```

Conjuntos (set)

```python
# Criando conjuntos
conjunto_vazio = set() # Importante: {} cria um dicionário vazio, use set() para conjunto vazio
numeros_unicos = {1, 2, 3, 4, 5, 1, 2} # Duplicados são automaticamente removidos
frutas_set = set(["maçã", "banana", "cereja", "maçã"])

print(f"\nConjunto de números únicos: {numeros_unicos}") # Saída: Conjunto de números únicos: {1, 2, 3, 4, 5}
print(f"Conjunto de frutas: {frutas_set}") # Saída: Conjunto de frutas: {'cereja', 'maçã', 'banana'} (ordem pode variar)
print(f"Tipo da variável 'numeros_unicos': {type(numeros_unicos)}")

# Adicionar elementos
frutas_set.add("uva")
print(f"Conjunto após adicionar 'uva': {frutas_set}") # Saída: Conjunto após adicionar 'uva': {'cereja', 'maçã', 'banana', 'uva'}

frutas_set.add("maçã") # Adicionar um item existente não faz nada
print(f"Conjunto após tentar adicionar 'maçã' novamente: {frutas_set}") # Saída: Conjunto após tentar adicionar 'maçã' novamente: {'cereja', 'maçã', 'banana', 'uva'}

# Remover elementos
frutas_set.remove("banana")
print(f"Conjunto após remover 'banana': {frutas_set}") # Saída: Conjunto após remover 'banana': {'cereja', 'maçã', 'uva'}

# Operações de Conjuntos
conjunto_a = {1, 2, 3, 4}
conjunto_b = {3, 4, 5, 6}

print(f"\nConjunto A: {conjunto_a}")
print(f"Conjunto B: {conjunto_b}")

print(f"União (elementos em A ou B): {conjunto_a.union(conjunto_b)}") # Saída: União (elementos em A ou B): {1, 2, 3, 4, 5, 6}
print(f"Interseção (elementos em A E B): {conjunto_a.intersection(conjunto_b)}") # Saída: Interseção (elementos em A E B): {3, 4}
print(f"Diferença (elementos em A mas NÃO em B): {conjunto_a.difference(conjunto_b)}") # Saída: Diferença (elementos em A mas NÃO em B): {1, 2}
print(f"É subconjunto? (A está contido em B): {conjunto_a.issubset(conjunto_b)}") # Saída: É subconjunto? (A está contido em B): False
```

Compreensão de Listas ([expressao for item in iteravel if condicao])

```python
# Criar uma lista de quadrados dos números de 0 a 9
quadrados = [x**2 for x in range(10)]
print(f"\nQuadrados de 0 a 9: {quadrados}") # Saída: Quadrados de 0 a 9: [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]

# Criar uma lista de números pares de 0 a 9
pares = [x for x in range(10) if x % 2 == 0]
print(f"Números pares de 0 a 9: {pares}") # Saída: Números pares de 0 a 9: [0, 2, 4, 6, 8]

# Converter strings para maiúsculas
nomes = ["ana", "bruno", "carla"]
nomes_maiusculos = [nome.upper() for nome in nomes]
print(f"Nomes em maiúsculas: {nomes_maiusculos}") # Saída: Nomes em maiúsculas: ['ANA', 'BRUNO', 'CARLA']
```

Compreensão de Dicionários ({chave_exp: valor_exp for item in iteravel if condicao})

```python
# Criar um dicionário com números e seus quadrados
dicionario_quadrados = {x: x**2 for x in range(5)}
print(f"\nDicionário de números e seus quadrados: {dicionario_quadrados}") # Saída: Dicionário de números e seus quadrados: {0: 0, 1: 1, 2: 4, 3: 9, 4: 16}

# Inverter chaves e valores de um dicionário
idades = {"Alice": 25, "Bob": 30, "Carlos": 25}
nomes_por_idade = {idade: nome for nome, idade in idades.items()}
print(f"Nomes por idade (chaves invertidas): {nomes_por_idade}") # Saída: Nomes por idade (chaves invertidas): {25: 'Carlos', 30: 'Bob'}
# Note que 'Alice' foi sobrescrita por 'Carlos' porque 25 já era chave. Conjuntos não aceitam duplicados.
```
